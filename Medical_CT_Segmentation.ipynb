```python
import nibabel as nib
import numpy as np
import plotly.graph_objects as go
from skimage import measure
import os
import matplotlib.cm as cm # Import colormap

# Load the segmented data
segmentation_path = '/content/Heart/heartchambers_highres.nii'

# Check if the file exists
if not os.path.exists(segmentation_path):
    raise FileNotFoundError(f"Segmentation file not found at: {segmentation_path}")

segmentation_img = nib.load(segmentation_path)
segmentation_data = segmentation_img.get_fdata()
affine_matrix = segmentation_img.affine

# Get unique labels in the segmentation data (excluding background 0)
unique_labels = np.unique(segmentation_data)
unique_labels = unique_labels[unique_labels > 0]

if len(unique_labels) == 0:
    raise ValueError("No segmented structures found in the data (only background).")

# Define a colormap for different labels
# Using a perceptually uniform colormap for better differentiation
colorscale = cm.get_cmap('viridis', len(unique_labels))
colors = [f'rgb({int(c[0]*255)}, {int(c[1]*255)}, {int(c[2]*255)})' for c in colorscale.colors]

# Define the list of potential heart segmentation files (needed to map labels to names)
output_dir = 'heartchambers_highres' # Define output_dir as it's used below
heart_segmentation_files = [
    'myocardium.nii.gz', 'atrium_left.nii.gz', 'ventricle_left.nii.gz',
    'atrium_right.nii.gz', 'ventricle_right.nii.gz', 'aorta.nii.gz',
    'pulmonary_artery.nii.gz', 'atrial_appendage_left.nii.gz'
]

# Filter for files that exist in the output directory (needed to map labels to names)
existing_heart_segmentation_files = [f for f in heart_segmentation_files if os.path.exists(os.path.join('/content', output_dir, f))]

# Initialize an empty list to store Plotly go.Mesh3d objects
data = [] # Renamed from mesh_data_list for clarity and consistency
label_names = {} # Dictionary to store label names (optional, based on file names if available)

# Iterate through each unique label and create a mesh for it
for i, label in enumerate(unique_labels):
    # Create a mask for the current label
    mask_data = (segmentation_data == label).astype(np.uint8)

    # Perform marching cubes to get vertices, faces, normals, values
    # Adjusting level slightly if needed, but 0.5 is standard for binary masks
    verts, faces, normals, values = measure.marching_cubes(mask_data, level=0.5)

    if verts.shape[0] > 0: # Only create mesh if surface extraction was successful
        # Apply affine transformation to vertices
        # Add a column of ones for the matrix multiplication
        verts_homogeneous = np.hstack((verts, np.ones((verts.shape[0], 1))))
        transformed_verts_homogeneous = verts_homogeneous @ affine_matrix.T
        transformed_verts = transformed_verts_homogeneous[:, :3]

        # Get the color for the current label
        if label == 5:
            mesh_color = 'red' # Set color to red for Label 5
        else:
            mesh_color = colors[i % len(colors)] # Use colormap for other labels


        # Create a Plotly mesh3d object
        # Assign a default opacity
        mesh = go.Mesh3d(
            x=transformed_verts[:, 0],
            y=transformed_verts[:, 1],
            z=transformed_verts[:, 2],
            i=faces[:, 0],
            j=faces[:, 1],
            k=faces[:, 2],
            name=f'Label {int(label)}', # Name for legend based on label value
            opacity=0.7, # Default opacity
            color=mesh_color # Set color based on the color logic
        )
        data.append(mesh)
        # Populate label_names with generic Label names
        label_names[int(label)] = f'Label {int(label)}'


# Fix: Check the 'data' list, not 'mesh_data_list'
if not data:
    raise ValueError("No mesh data was generated for any label.")

# Store the list of mesh data and labels for the Dash app
# In a real Dash app, these would be passed or accessed in callbacks
# For this subtask, just ensure they are available as variables
# data (now renamed from mesh_data_list) and unique_labels are now available
mesh_data_list = data # Assign back to mesh_data_list for consistency with Dash layout cell
```

```python
def calculate_dice_score(ground_truth, predicted):
    """Calculates the Dice score between two binary segmentation masks."""
    intersection = np.sum(ground_truth * predicted)
    sum_of_volumes = np.sum(ground_truth) + np.sum(predicted)
    if sum_of_volumes == 0:
        return 1.0 # Handle the case where both are empty
    return (2.0 * intersection) / sum_of_volumes

def calculate_iou(ground_truth, predicted):
    """Calculates the Intersection over Union (IoU) between two binary segmentation masks."""
    intersection = np.sum(ground_truth * predicted)
    union = np.sum(ground_truth) + np.sum(predicted) - intersection
    if union == 0:
        return 1.0 # Handle the case where both are empty
    return intersection / union

def calculate_volumetric_similarity(ground_truth, predicted):
    """Calculates the Volumetric Similarity between two binary segmentation masks."""
    volume_ground_truth = np.sum(ground_truth)
    volume_predicted = np.sum(predicted)
    if (volume_ground_truth + volume_predicted) == 0:
        return 1.0 # Handle the case where both are empty
    return 1.0 - abs(volume_ground_truth - volume_predicted) / (volume_ground_truth + volume_predicted)

# Compute the metrics using the loaded data
dice_score = calculate_dice_score(ground_truth_data, predicted_data)
iou = calculate_iou(ground_truth_data, predicted_data)
volumetric_similarity = calculate_volumetric_similarity(ground_truth_data, predicted_data)

print(f"Dice Score: {dice_score}")
print(f"IoU: {iou}")
print(f"Volumetric Similarity: {volumetric_similarity}")
```

## Calculate evaluation metrics for each lung label

### Subtask:
Iterate through each identified lung label, load its predicted and ground truth segmentation data, and calculate the Dice score, IoU, and Volumetric Similarity for each label.


**Reasoning**:
Initialize a dictionary to store lung metrics and iterate through the identified lung segmentation paths to calculate metrics for each.



```python
# Initialize an empty dictionary to store metrics for each lung label
lung_metrics = {}

# Iterate through the list of identified lung segmentation file paths
for predicted_path in lung_segmentation_paths:
    # Extract the filename and label name from the predicted path
    filename = os.path.basename(predicted_path)
    label_name = filename.replace('.nii.gz', '') # Use filename without extension as label name

    # Construct the corresponding ground truth file path
    ground_truth_path = predicted_path.replace('/content/s0011_seg/', '/content/s0011/segmentations/')

    # Check if the ground truth file exists
    if not os.path.exists(ground_truth_path):
        print(f"Warning: Ground truth file not found for {filename}. Skipping.")
        continue

    try:
        # Load the predicted NIfTI image and get the data
        predicted_img = nib.load(predicted_path)
        predicted_data = predicted_img.get_fdata()

        # Load the corresponding ground truth NIfTI image and get the data
        ground_truth_img = nib.load(ground_truth_path)
        ground_truth_data = ground_truth_img.get_fdata()

        # Ensure both arrays have the same shape
        if predicted_data.shape != ground_truth_data.shape:
            print(f"Warning: Shape mismatch for {filename}. Predicted: {predicted_data.shape}, Ground Truth: {ground_truth_data.shape}. Skipping.")
            continue

        # Calculate the metrics
        dice_score = calculate_dice_score(ground_truth_data, predicted_data)
        iou = calculate_iou(ground_truth_data, predicted_data)
        volumetric_similarity = calculate_volumetric_similarity(ground_truth_data, predicted_data)

        # Store the calculated metrics in the dictionary
        lung_metrics[label_name] = {
            'Dice Score': dice_score,
            'IoU': iou,
            'Volumetric Similarity': volumetric_similarity
        }

    except Exception as e:
        print(f"Error processing {filename}: {e}. Skipping.")
        continue

# Print the calculated metrics for each lung label
print("\nCalculated Metrics for Lung Segmentations:")
for label, metrics in lung_metrics.items():
    print(f"\n{label}:")
    for metric_name, value in metrics.items():
        print(f"  {metric_name}: {value:.4f}")
```

## Prepare data for visualization

### Subtask:
Load the segmented lung data, extract the unique labels, and generate mesh data for each label using marching cubes.


**Reasoning**:
Load the individual lung segmentation files, combine them into a single NIfTI image with distinct labels for each lobe, and then proceed to extract unique labels and generate mesh data for each labeled region, applying the affine transformation to the vertices.



```python
import nibabel as nib
import numpy as np
import plotly.graph_objects as go
from skimage import measure
import os
import matplotlib.cm as cm

# Assuming lung_segmentation_paths contains the paths to individual lung lobe segmentations
# Example: ['/content/s0011_seg/lung_lower_lobe_right.nii.gz', ...]

if not lung_segmentation_paths:
    raise ValueError("No lung segmentation files found to process.")

# Load the first image to get dimensions and affine matrix
first_img = nib.load(lung_segmentation_paths[0])
image_shape = first_img.shape
affine_matrix = first_img.affine

# Initialize a combined segmentation data array with zeros
combined_segmentation_data = np.zeros(image_shape, dtype=np.uint8)

# Define a dictionary to map filenames to integer labels
label_map = {
    'lung_lower_lobe_right.nii.gz': 1,
    'lung_middle_lobe_right.nii.gz': 2,
    'lung_lower_lobe_left.nii.gz': 3,
    'lung_upper_lobe_left.nii.gz': 4,
    'lung_upper_lobe_right.nii.gz': 5
}

# Iterate through each lung segmentation file and combine them into a single volume
# Assign a unique integer label to each lobe
for predicted_path in lung_segmentation_paths:
    filename = os.path.basename(predicted_path)
    if filename in label_map:
        label = label_map[filename]
        try:
            img = nib.load(predicted_path)
            data = img.get_fdata()

            # Ensure data shape matches the first image
            if data.shape != image_shape:
                print(f"Warning: Shape mismatch for {filename}. Expected: {image_shape}, Got: {data.shape}. Skipping.")
                continue

            # Add the current lobe's segmentation to the combined data with its unique label
            # Use np.logical_or to handle potential overlaps (though ideally lobes shouldn't overlap significantly)
            # Or simply assign the label where the data is > 0
            combined_segmentation_data[data > 0] = label

        except Exception as e:
            print(f"Error loading or processing {filename}: {e}. Skipping.")
            continue
    else:
        print(f"Warning: Unknown lung segmentation file: {filename}. Skipping.")


# Get unique labels in the combined segmentation data (excluding background 0)
unique_labels = np.unique(combined_segmentation_data)
unique_labels = unique_labels[unique_labels > 0]

if len(unique_labels) == 0:
    raise ValueError("No segmented lung structures found in the combined data.")

# Define a colormap for different labels
# Using a perceptually uniform colormap for better differentiation
colorscale = cm.get_cmap('viridis', len(unique_labels))
colors = [f'rgb({int(c[0]*255)}, {int(c[1]*255)}, {int(c[2]*255)})' for c in colorscale.colors]

# Initialize an empty list to store Plotly go.Mesh3d objects
mesh_data_list = []
label_names = {v: k.replace('.nii.gz', '') for k, v in label_map.items()} # Map integer labels back to names

# Iterate through each unique label and create a mesh for it
for i, label in enumerate(unique_labels):
    # Create a mask for the current label
    mask_data = (combined_segmentation_data == label).astype(np.uint8)

    # Perform marching cubes to get vertices, faces, normals, values
    # Adjusting level slightly if needed, but 0.5 is standard for binary masks
    try:
        verts, faces, normals, values = measure.marching_cubes(mask_data, level=0.5)
    except Exception as e:
         print(f"Error during marching cubes for label {label}: {e}. Skipping mesh generation for this label.")
         continue


    if verts.shape[0] > 0: # Only create mesh if surface extraction was successful
        # Apply affine transformation to vertices
        # Add a column of ones for the matrix multiplication
        verts_homogeneous = np.hstack((verts, np.ones((verts.shape[0], 1))))
        transformed_verts_homogeneous = verts_homogeneous @ affine_matrix.T
        transformed_verts = transformed_verts_homogeneous[:, :3]

        # Get the color for the current label
        mesh_color = colors[i % len(colors)] # Use colormap for other labels

        # Get the name for the current label
        mesh_name = label_names.get(label, f'Label {int(label)}')


        # Create a Plotly mesh3d object
        # Assign a default opacity
        mesh = go.Mesh3d(
            x=transformed_verts[:, 0],
            y=transformed_verts[:, 1],
            z=transformed_verts[:, 2],
            i=faces[:, 0],
            j=faces[:, 1],
            k=faces[:, 2],
            name=mesh_name, # Name for legend based on label name
            opacity=0.7, # Default opacity
            color=mesh_color # Set color based on the color logic
        )
        mesh_data_list.append(mesh)


# Check if any mesh data was generated
if not mesh_data_list:
    raise ValueError("No mesh data was generated for any lung label.")

# mesh_data_list and label_names are now available for the next step
```

## Prepare data for visualization

### Subtask:
Load the segmented vertebrae data, extract the unique labels, and generate mesh data for each label using marching cubes.


**Reasoning**:
Load the first vertebrae segmentation image to get dimensions and affine matrix, initialize a combined segmentation data array, define a label map, iterate through segmentation paths to combine data with unique labels, get unique labels, define a colormap, initialize mesh data list and label names dictionary, iterate through unique labels to create masks, perform marching cubes, transform vertices, get color and name, create Plotly mesh3d objects, and check if any mesh data was generated.



```python
import numpy as np
import plotly.graph_objects as go
from skimage import measure
import os
import matplotlib.cm as cm

# Assuming vertebrae_segmentation_paths contains the paths to individual vertebrae segmentations
if not vertebrae_segmentation_paths:
    raise ValueError("No vertebrae segmentation files found to process.")

# Load the first image to get dimensions and affine matrix
first_img = nib.load(vertebrae_segmentation_paths[0])
image_shape = first_img.shape
affine_matrix = first_img.affine

# Initialize a combined segmentation data array with zeros
combined_segmentation_data = np.zeros(image_shape, dtype=np.uint8)

# Define a dictionary to map filenames to integer labels
label_map = {}
# Populate label_map based on the identified vertebrae segmentation files
for i, predicted_path in enumerate(vertebrae_segmentation_paths):
    filename = os.path.basename(predicted_path)
    label_map[filename] = i + 1 # Assign unique integer label starting from 1

# Iterate through each vertebrae segmentation file and combine them into a single volume
# Assign a unique integer label to each vertebrae
for predicted_path in vertebrae_segmentation_paths:
    filename = os.path.basename(predicted_path)
    if filename in label_map:
        label = label_map[filename]
        try:
            img = nib.load(predicted_path)
            data = img.get_fdata()

            # Ensure data shape matches the first image
            if data.shape != image_shape:
                print(f"Warning: Shape mismatch for {filename}. Expected: {image_shape}, Got: {data.shape}. Skipping.")
                continue

            # Assign the current vertebrae's segmentation to the combined data with its unique label
            # Only assign if the voxel is part of the segmentation (data > 0) and hasn't been assigned yet
            # This handles potential overlaps by prioritizing the first encountered label (based on iteration order)
            # A more robust approach for overlaps might be needed depending on the data
            combined_segmentation_data[(data > 0) & (combined_segmentation_data == 0)] = label


        except Exception as e:
            print(f"Error loading or processing {filename}: {e}. Skipping.")
            continue
    else:
        print(f"Warning: Unknown vertebrae segmentation file: {filename}. Skipping.")


# Get unique labels in the combined segmentation data (excluding background 0)
unique_labels = np.unique(combined_segmentation_data)
unique_labels = unique_labels[unique_labels > 0]

if len(unique_labels) == 0:
    raise ValueError("No segmented vertebrae structures found in the combined data.")

# Define a colormap for different labels
# Using a perceptually uniform colormap for better differentiation
colorscale = cm.get_cmap('viridis', len(unique_labels))
colors = [f'rgb({int(c[0]*255)}, {int(c[1]*255)}, {int(c[2]*255)})' for c in colorscale.colors]

# Initialize an empty list to store Plotly go.Mesh3d objects
mesh_data_list = []
# Create a dictionary to map integer labels back to their corresponding vertebrae names
label_names = {v: k.replace('.nii.gz', '') for k, v in label_map.items()}

# Iterate through each unique label and create a mesh for it
for i, label in enumerate(unique_labels):
    # Create a mask for the current label
    mask_data = (combined_segmentation_data == label).astype(np.uint8)

    # Perform marching cubes to get vertices, faces, normals, values
    # Adjusting level slightly if needed, but 0.5 is standard for binary masks
    try:
        verts, faces, normals, values = measure.marching_cubes(mask_data, level=0.5)
    except Exception as e:
         print(f"Error during marching cubes for label {label}: {e}. Skipping mesh generation for this label.")
         continue


    if verts.shape[0] > 0: # Only create mesh if surface extraction was successful
        # Apply affine transformation to vertices
        # Add a column of ones for the matrix multiplication
        verts_homogeneous = np.hstack((verts, np.ones((verts.shape[0], 1))))
        transformed_verts_homogeneous = verts_homogeneous @ affine_matrix.T
        transformed_verts = transformed_verts_homogeneous[:, :3]

        # Get the color for the current label
        mesh_color = colors[i % len(colors)] # Use colormap for other labels

        # Get the name for the current label
        mesh_name = label_names.get(label, f'Label {int(label)}')

        # Create a Plotly mesh3d object
        # Assign a default opacity
        mesh = go.Mesh3d(
            x=transformed_verts[:, 0],
            y=transformed_verts[:, 1],
            z=transformed_verts[:, 2],
            i=faces[:, 0],
            j=faces[:, 1],
            k=faces[:, 2],
            name=mesh_name, # Name for legend based on label name
            opacity=0.7, # Default opacity
            color=mesh_color # Set color based on the color logic
        )
        mesh_data_list.append(mesh)


# Check if any mesh data was generated
if not mesh_data_list:
    raise ValueError("No mesh data was generated for any vertebrae label.")

# mesh_data_list and label_names are now available for the next step
```
